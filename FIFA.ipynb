# 4. Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix.

```python
df['Value'].dtype
```

```python
df['Wage'].dtype
```

```python
df['Release Clause'].dtype
```

```python
df['Value'] = df['Value'].replace({'\€': ''}, regex=True)
df['Wage'] = df['Wage'].replace({'\€': ''}, regex=True)
df['Release Clause'] = df['Release Clause'].replace({'\€': ''}, regex=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
def currency_convert(x):
    if pd.isnull(x) == True:
        return float(x)
    if 'M' in x:
        return float(x.replace('M', ''))*1000000
    elif 'K' in x:
        return float(x.replace('K', ''))*1000
    else:
        return float(x)
        
df['Value'] = df.apply(lambda x: currency_convert(x['Value']),axis=1)
```

```python
df['Wage'] = df.apply(lambda x: currency_convert(x['Wage']),axis=1)
```

```python
df['Release Clause'] = df.apply(lambda x: currency_convert(x['Release Clause']),axis=1)
```

```python
df
```

# 7. The column 'Height' is in inches with a quotation mark, Convert to float with decimal points

```python
df['Height']
```

```python
def parse_ht(x):
    # format: 7' 0.0"
    x=str(x)
    ht_ = x.split("'")
    ft_ = float(ht_[0])
    if len(ht_)>1:
        in_ = float(ht_[1])
    else:
        in_=0.0
    return ft_ + (in_/12.0)
    

df['Height']=df.apply(lambda x: parse_ht(x['Height']),axis=1)
```
